```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.multivariate_normal([1, 5], [[1, 0.3],[0.3, 1]], size=5000)
x, y = data[:,0], data[:,1]
plt.scatter(x, y, s=1, c='red');
```
